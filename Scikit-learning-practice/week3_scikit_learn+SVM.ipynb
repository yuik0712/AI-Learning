# 데이터 셋 읽기 (iris)


```python
from sklearn import datasets # sklearn 모듈 사용하여 datasets 클래스 불러옴

d = datasets.load_iris() # load_iris() 함수 호출 -> 객체 d에 저장
print(d.DESCR) # 객체 d의 DESCR 변수 출력
```

```python
# 내용 확인
for i in range(0, len(d.data)):
  print(i + 1, d.data[i], d.target[i])
```

# SVM 학습 모델

```python
from sklearn import svm

s = svm.SVC(gamma=0.1, C=10) # svm 분류기 모델 svc 객체 생성하여 s에 저장
s.fit(d.data, d.target) # 훈련 집합을 가지고 학습 수행 (매개변수 특징 -> data, target으로 설정)

new_d = [[6.4, 3.2, 6.0, 2.5], [7.1, 3.1, 4.7, 1.35]] # 샘플을 변형하여 새로운 데이터 생성

res = s.predict(new_d) # 테스트 집합을 가지고 예측 수행

print("새로운 2개 샘플의 부류", res) # 결과 출력
```

# 훈련/검증/테스트 집합으로 쪼개기

```python
from sklearn import datasets
from sklearn import svm
from sklearn.model_selection import train_test_split
import numpy as np

# 데이터셋을 읽고 훈련 집합과 테스트 집합으로 분할
digit = datasets.load_digits()
x_train, x_test, y_train, y_test = train_test_split(digital.data, digit.target, test_size=0.6) # train_test_split 함수로 훈련 60% (나머지 40% -> 테스트) 랜덤 분할

# svm의 분류 모델 SVC 학습
s = svm.SVC(gamma=0.001)
s.fit(x_train, y_train) # 훈련집합 -> fit 함수에 주어 학습

res = s.predict(x_test) # 테스트 집합의 특징 벡터 -> predict 함수에 주어 예측 수행

# 테스트 집합의 레이블 y_test 혼돈 행렬 계산
conf = np.zeros((10, 10))
for i in range(len(res)):
  conf[res[i][y_test[i]] += 1]
print(conf)
```

```python
# 정확률 측정 및 출력
no_correct = 0
for i in range(10):
  no_correct += conf[i][i]
accuracy = no_correct / len(res)
print("정확률은", accuracy * 100, "%입니다.")
```

# 교차검증(digit 데이터)

```python
from sklearn import datasets
from sklearn import svm
from sklearn.model_selection import cross_val_score
import numpy as np

digit = datasets.load_digits()
s = svm.SVC(gamma=0.001)
accuracies = cross_val_score(s, digit.data, digit.target, cv=5) # 5겹 교차 검증

print(accuracies)
print("정확률(평균)=%0.3f, 표준편차=0.3f"%(accuracies.mean()*100, accuracies.std()))
```
